```python
#we then combine all of the genres, movies, and ratings
#full_movies = pd.concat([movies,genre_list], axis = 1)
dataset = pd.merge(movies, ratings)
dataset = dataset.dropna()

#find the mean of the ratings for each movie
ratings_mean = (dataset.groupby('title'))['title','rating'].mean()
ratings_mean['title'] = ratings_mean.index
ratings_mean = pd.DataFrame({'title': ratings_mean['title'], 'mean rating': ratings_mean['rating']})

#find the total number of ratings for each movie
ratings_total = dataset.groupby('title').size()
ratings_total = pd.DataFrame({'title':ratings_total.index,'total ratings': ratings_total.values})

#combine them all to create a dataset
dataset_clean = pd.merge(ratings_mean, pd.merge(movies,ratings_total)).sort_values(by = 'total ratings',ascending= False)
dataset_clean
```

```python
# This is a simple recommendation for the highest rated movies that are relatively popular
# It is useful for predicting 

simp_pred = dataset_clean.drop(dataset_clean[dataset_clean["total ratings"] < 1000].index).sort_values(by = 'mean rating',ascending= False)

simp_pred
```

```python
#This is a content specific recommender that weighs the movies that you like and ranks them accordingly

my_ratings = dataset.drop(dataset[dataset["userId"] != 138494].index)

movies_watched = []
for movie in my_ratings.title:
    movies_watched.append(movie)
    
def generate_user_data():
    user_data = {
        '(no genres listed)': 0,
        'Action': 0,
        'Adventure': 0,
        'Animation': 0,
        'Children': 0,
        'Comedy': 0,
        'Crime': 0,
        'Documentary': 0,
        'Drama': 0,
        'Fantasy': 0,
        'Film-Noir': 0,
        'Horror': 0,
        'IMAX': 0,
        'Musical': 0,
        'Mystery': 0,
        'Romance': 0,
        'Sci-Fi': 0,
        'Thriller': 0,
        'War': 0,
        'Western': 0
    }

    user_num = {
        '(no genres listed)': 0,
        'Action': 0,
        'Adventure': 0,
        'Animation': 0,
        'Children': 0,
        'Comedy': 0,
        'Crime': 0,
        'Documentary': 0,
        'Drama': 0,
        'Fantasy': 0,
        'Film-Noir': 0,
        'Horror': 0,
        'IMAX': 0,
        'Musical': 0,
        'Mystery': 0,
        'Romance': 0,
        'Sci-Fi': 0,
        'Thriller': 0,
        'War': 0,
        'Western': 0
    }

    for index, movie in my_ratings.iterrows():
        split = movie["genres"].split("|")
        for text in split:
            user_data[text] += movie["rating"]/5
            user_num[text]+=1
        
    if(user_num["(no genres listed)"]>0):
        user_data["(no genres listed)"]/=user_num["(no genres listed)"]
    if(user_num["Action"]>0):
        user_data["Action"]/=user_num["Action"]
    if(user_num["Adventure"]>0):
        user_data["Adventure"]/=user_num["Adventure"]
    if(user_num["Animation"]>0):
        user_data["Animation"]/=user_num["Animation"]
    if(user_num["Children"]>0):
        user_data["Children"]/=user_num["Children"]
    if(user_num["Comedy"]>0):
        user_data["Comedy"]/=user_num["Comedy"]
    if(user_num["Crime"]>0):
        user_data["Crime"]/=user_num["Crime"]
    if(user_num["Documentary"]>0):
        user_data["Documentary"]/=user_num["Documentary"]
    if(user_num["Drama"]>0):
        user_data["Drama"]/=user_num["Drama"]
    if(user_num["Fantasy"]>0):
        user_data["Fantasy"]/=user_num["Fantasy"]
    if(user_num["Film-Noir"]>0):
        user_data["Film-Noir"]/=user_num["Film-Noir"]
    if(user_num["Horror"]>0):
        user_data["Horror"]/=user_num["Horror"]
    if(user_num["IMAX"]>0):
        user_data["IMAX"]/=user_num["IMAX"]
    if(user_num["Musical"]>0):
        user_data["Musical"]/=user_num["Musical"]
    if(user_num["Mystery"]>0):
        user_data["Mystery"]/=user_num["Mystery"]
    if(user_num["Romance"]>0):
        user_data["Romance"]/=user_num["Romance"]
    if(user_num["Sci-Fi"]>0):
        user_data["Sci-Fi"]/=user_num["Sci-Fi"]
    if(user_num["Thriller"]>0):
        user_data["Thriller"]/=user_num["Thriller"]
    if(user_num["War"]>0):
        user_data["War"]/=user_num["War"]
    if(user_num["Western"]>0):
        user_data["Western"]/=user_num["Western"]
    return user_data

user_data = generate_user_data()

user_data
```

```python
#generating recommendations
def recommend(user_data):
    personal_list = pd.DataFrame({
        'personal_score' : []
    })
    recommendations = pd.concat([simp_pred,personal_list], axis = 1)
    
    for index, movie in recommendations.iterrows():
        personal_data = pd.DataFrame({
            'personal_score' : [0] 
        })
        personal_total = 0
        genre_total = 0
        split = movie['genres'].split("|")
        for text in split:
            personal_total += user_data[text]
            genre_total += 1
        recommendations.at[index, 'personal_score'] = (personal_total/genre_total)*movie['mean rating']

    recommendations = recommendations.sort_values(by = 'personal_score',ascending= False)

    for movie in movies_watched:
        recommendations = recommendations.drop(recommendations[recommendations["title"] == movie].index)

    return recommendations

recommend(user_data).head(10)
```

```python
#generating recommendations
def recommend_game(user_data):
    personal_list = pd.DataFrame({
        'personal_score' : []
    })
    recommendations = pd.concat([games,personal_list], axis = 1)
    
    for index, game in recommendations.iterrows():
        personal_data = pd.DataFrame({
            'personal_score' : [0] 
        })
        personal_total = 0
        genre_total = 0
        split = game['Genre'].split(", ")
        for text in split:
            personal_total += user_data[text]
            genre_total += 1
        recommendations.at[index, 'personal_score'] = (personal_total/genre_total)*game['Score']

    recommendations = recommendations.sort_values(by = 'personal_score',ascending= False)

    for game in games_played:
        recommendations = recommendations.drop(recommendations[recommendations["Game"] == game].index)

    return recommendations

recommend_game(user_game_data).head(10)
```
